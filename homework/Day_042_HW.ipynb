```python
from sklearn import datasets, metrics # mean_squared_error, r2_score, accuracy_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

## wine

```python
wine = datasets.load_wine()
wine
```

```python
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=4)
```

```python
list(zip(x_train, y_train))
```

```python
list(zip(x_test, y_test))
```

```python
# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)

y_pred
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
print(wine.feature_names)
```

```python
print("Feature importance: ", clf.feature_importances_)
```

```python
sorted(list(zip(wine.feature_names, clf.feature_importances_)), reverse = True ,key = lambda x:x[1])
```

## boston

```python
boston = datasets.load_boston()
boston
```

```python
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.25, random_state=4)

# 建立模型
clf = DecisionTreeRegressor()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)

y_pred
```

```python
mean_squared_error = metrics.mean_squared_error(y_test, y_pred)
print("mean_squared_error: ", mean_squared_error)
```

```python
r2_score = metrics.r2_score(y_test, y_pred)
print("mean_squared_error: ", r2_score)
```
